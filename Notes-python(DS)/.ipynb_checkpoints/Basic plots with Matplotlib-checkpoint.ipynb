```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263,6.972]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8],
          ['0', '2B', '4B', '6B']) #input is list
plt.text(1970, 3.692, 'Text')#Additional Customization
plt.grid(True)

plt.show() #display the plot
```

```python
plt.scatter(year, pop)
plt.show()
```

#### The number of `bins` is pretty important. Too few bins will oversimplify reality and won't show you the details. Too many bins will overcomplicate reality and won't show the bigger picture.  
#### `plt.show()` displays a plot; `plt.clf()` cleans it up again so you can start afresh.

```python
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins=3)
plt.show()
plt.clf()

plt.hist(values, bins=5)
plt.show()
plt.clf()
```
